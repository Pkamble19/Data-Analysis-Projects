##### Importing Required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

##### Importing csv file

```python
files=os.listdir('Sales_Data')
path=r'C:\Users\sandy4232\Desktop\Data analyst projects\Sales Data Analysis\Sales_Data'
sales_data=pd.DataFrame()
```

```python
for data in files :
    sales_data_sa=pd.read_csv(path + "\\"+ data)
    sales_data=pd.concat([sales_data,sales_data_sa])
sales_data.shape
```

```python
sales_data
```

#### cheacking for Empty rows    

```python
sales_data.dropna(inplace=True)
sales_data.reset_index(inplace=True)
sales_data.drop('index',axis=1,inplace=True)
```

##### Cleaning data

```python
sales_data[sales_data['Order ID']=='Order ID']
```

```python
sales_data.dtypes
sales_data.drop(index=sales_data[sales_data['Order ID']=='Order ID'].index,inplace=True)
```

```python
sales_data
```

```python
def month(x):
    return x.split('/')[0]
def day(x):
    return x.split('/')[1]
def year(x):
    return x.split('/')[2].split(' ')[0]
```

```python
sales_data['Month']=sales_data['Order Date'].apply(month)
```

```python
sales_data['Hour']=pd.to_datetime(sales_data['Order Date']).dt.hour
```

```python
sales_data
```

##### seting data types for data

```python
sales_data['Order Date']=pd.to_datetime(sales_data['Order Date'])
sales_data['Order ID']=pd.to_numeric(sales_data['Order ID'])
sales_data['Price Each']=pd.to_numeric(sales_data['Price Each'])
sales_data['Quantity Ordered']=pd.to_numeric(sales_data['Quantity Ordered'])
sales_data['Month']=pd.to_numeric(sales_data['Month'])
```

```python
sales_data.dtypes
```

##### Saving new combined cleaned csv

#### FINDING WHICH CITY HAS MAX ORDERS

```python
def city(x):
    return x.split(',')[1]
    
```

```python
sales_data['CITY']=sales_data['Purchase Address'].apply(city)
```

```python
orde_city=sales_data['CITY'].value_counts().reset_index()
orde_city.columns=['City','No Of Orders']
```

```python
#Top 5 city with maximum orders
orde_city.head()
```

```python
plt.figure(figsize=(12,6))
plt.bar('City','No Of Orders',data=orde_city)
plt.xlabel('CITY')
plt.ylabel('Numbers Of Orders')
```
