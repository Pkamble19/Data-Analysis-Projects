```python
def perform_analysis(data,feature,col1,col2,chart):
    '''This will return us some plots depending upon whatever chart we want.
    
    perform_analysis(data,feature,col1,col2,chart)
    
    
    parameters
    ----------
    data:dataframe
    
    fetures:column_name that we want to consider for analysis
    
    col1:1st column name that you want to asign to new df
    
    col2:1st column name that you want to asign to new df
    
    chart :if value is 1 ,it represent bar chart
    
    chart :if value is 1 ,it represent pie chart
    
    '''
    dataframe=data[feature].value_counts().reset_index()
    dataframe.columns=[col1,col2]
    print('Top 5\n',dataframe.head())
    if chart==1:
        plt.bar(dataframe[col1][0:10],dataframe[col2][0:10])
        plt.xticks(rotation='vertical')
    elif chart==2:
            plt.figure(figsize=(12,6))
            plt.pie(labels=dataframe[col1][0:10],x=dataframe[col2][0:10])
            plt.xticks()
            
        
```

```python
perform_analysis(data,'company','Company_name','Number of jobs',1)
```

```python
perform_analysis(data,'company','Company_name','Number of jobs',2)
```
