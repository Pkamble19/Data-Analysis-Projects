#data cleaning for payrate column


```python
#we have to seprate this values in min pay rate and max payrate 
naukridata['payrate']
```

```python
payrate_len=[len(str(pay).split('-')) for pay in naukridata['payrate']]
```

```python
pd.Series(payrate_len).value_counts()
```

```python
payrate_split=naukridata['payrate'].str.split('-',expand=True)
payrate_split
```

```python
payrate_split[0]=payrate_split[0].str.replace(',','')
payrate_split[0]=payrate_split[0].str.strip()
```

```python
payrate_split
```

```python
pattern='\D.*'
```

```python
payrate_split[0]=payrate_split[0].str.replace(pattern,'')
```

```python
payrate_split
```

```python
payrate_split[1]=payrate_split[1].str.strip()
```

```python
payrate_split[1]=payrate_split[1].apply(lambda x:str(x).replace(',',''))
```

```python
payrate_split[1]
```

```python
payrate_split[1]=payrate_split[1].str.replace(pattern,'')
```

```python
payrate_split[1]
```

```python
payrate_split.dtypes
```

```python
payrate_split[0]=pd.to_numeric(payrate_split[0],errors='coerce')
payrate_split[1]=pd.to_numeric(payrate_split[1],errors='coerce')
```

```python
payrate_split.dtypes
```

```python
pay=pd.concat([payrate_split[0],payrate_split[1]],axis=1,sort=False)
pay.columns=['min_pay','max_pay']
```

```python
naukridata=pd.concat([naukridata,pay],axis=1,sort=False)
```

```python
naukridata.drop('payrate',axis=1,inplace=True)
```

```python
naukridata
```

```python
naukridata['avg_payrate']=(naukridata['min_pay']+naukridata['max_pay'])/2
```

```python
naukridata
```

```python
def fetch_dt_att(dataframe,feature):
    try:
        return pd.Series([dataframe[feature].dt.day,dataframe[feature].dt.month,dataframe[feature].dt.year])
    except:
        print('dose not support')
                                                     
```

```python
naukridata[['day','month','year']]=fetch_dt_att(naukridata,'postdate')
```
